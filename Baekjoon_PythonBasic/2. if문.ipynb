### 2-1 두 수 비교하기

<img src = "https://drive.google.com/uc?id=1-RZNdVkZ3IR6gpb_HsOG9TP9eAANFvBh" height = 70% width = 70%>

```python
a, b = map(int, input().split())

if a > b :
    print('>')
elif a < b :
    print('<')
elif a == b :
    print('==')
```

### 2-2 시험 성적

<img src = "https://drive.google.com/uc?id=1-RZNdVkZ3IR6gpb_HsOG9TP9eAANFvBh" height = 70% width = 70%>

```python
a = int(input())

if 0 <= a <= 100:
    if a >= 90:
        print('A')
    elif a >= 80:
        print('B')
    elif a >= 70:
        print('C')
    elif a >= 60:
        print('D')
    else:
        print('F')
```

### 2-3 윤년

<img src = "https://drive.google.com/uc?id=1-RZNdVkZ3IR6gpb_HsOG9TP9eAANFvBh" height = 70% width = 70%>

```python
a = int(input())

if 1 <= a <= 4000:
    if a % 4 == 0 and a % 100 != 0 or a % 400 == 0:
        print('1')
    else:
        print('0')
```

### 2-4 사분면 고르기

<img src = "https://drive.google.com/uc?id=1-RZNdVkZ3IR6gpb_HsOG9TP9eAANFvBh" height = 70% width = 70%>

```python
x = int(input())
y = int(input())

if (-1000 <= x <= 1000 and x != 0) and (-1000 <= y <= 1000 and y != 0) :
    if x > 0 and y > 0:
        print('1')
    elif x < 0 and y > 0:
        print('2')
    elif x < 0 and y < 0:
        print('3')
    else:
        print('4')
```

### 2-5 알람 시계

<img src = "https://drive.google.com/uc?id=1-RZNdVkZ3IR6gpb_HsOG9TP9eAANFvBh" height = 70% width = 70%>

```python
H, M = map(int, input().split())

if 0 <= H <= 23 and 0 <= M <= 59:
    if H == 0 and M >= 45:
        print(H, M - 45)
    elif H == 0 and M < 45:
        print(H + 24 - 1, M + 60 - 45)
    elif H != 0 and M >= 45:
        print(H, M - 45)
    elif H != 0 and M < 45:
        print(H - 1, M + 60 - 45)
```

### 2-7 주사위 세개

<img src = "https://drive.google.com/uc?id=1-RZNdVkZ3IR6gpb_HsOG9TP9eAANFvBh" height = 70% width = 70%>

```python
a, b, c = map(int,input().split())

if 0 <= a <= 6 and 0 <= b <= 6 and 0 <= c <= 6:
    if a == b == c:
        print(10000 + a * 1000)
    elif a == b or b == c:
        print(1000 + b * 100)
    elif a == c:
        print(1000 + a * 100)
    elif a != b and b != c and a != c:
        if a > b and a > c:
            print(a * 100)
        elif b > a and b > c:
            print(b * 100)
        elif c > a and c > a:
            print(c * 100)
```
