### 3-4 영수

<img src = "https://drive.google.com/uc?id=1W7Z1oefIYu1RFAUnARnnz40KUZN5dfoV" height = 100% width = 100%>

<img src = "https://drive.google.com/uc?id=1Heu6dgYC3RP2siraBB8I0bGzbsQx6HVI" height = 100% width = 100%>

<img src = "https://drive.google.com/uc?id=11fhX7SnjXbzcax5o6xMtrwY6387w-3fA" height = 100% width = 100%>

```python
x = int(input())    # 총 금액
n = int(input())    # 물건의 종류 수

resultX = 0

if 1 <= x <= 1000000000 and 1 <= n <= 100:
    for i in range(n):
        a, b = map(int, input().split())
        resultX += a * b     # result = result + a * b
    if 1 <= a <= 1000000 and 1 <= b <= 10:
        if x == resultX:
            print('Yes')
        elif x != resultX:
            print('No')
```

### 3-5 코딩은 체육과목입니다.

<img src = "https://drive.google.com/uc?id=1kVT9PGaFx5lFBMFdd92M2vxygd5gm3-9" height = 100% width = 100%>

<img src = "https://drive.google.com/uc?id=1GzVutSN5MRgieTZNqcD9AprHQDVkwicG" height = 100% width = 100%>

```python
n = int(input())

x = ''

if 4 <= n <= 1000 and n % 4 == 0:
    if n == 4:
        print('long int')
    elif n > 4:
        for i in range(n // 4):
            x += 'long '     # x = 'long' + 'long'
        print(x + 'int')
```

### 3-6 빠른 A+B

<img src = "https://drive.google.com/uc?id=1umw1modYwGqOXQy2JMpxsZMdHZXsRMVI" height = 100% width = 100%>

<img src = "https://drive.google.com/uc?id=1BFf32cWkVVBAOCCtvCwwcKoPiI_oHs60" height = 100% width = 100%>

```python
import sys
from io import StringIO

# 테스트용 입력 데이터 (여러 줄 입력)
input_data = """5
1 1
12 34
5 500
40 60
1000 1000
"""

# StringIO를 사용하여 sys.stdin 대체
sys.stdin = StringIO(input_data)

input = sys.stdin.readline  # 빠른 입력을 StringIO로 대체
t = int(input().strip())    # 첫 줄에서 테스트 케이스 개수 입력

for _ in range(t):
    n1, n2 = map(int, input().split())  # 두 수 입력받기
    print(n1 + n2)                     # 두 수의 합 출력
```

### 3-10 별 찍기 - 2

<img src = "https://drive.google.com/uc?id=1_QCZ9W6kEKR64bH89SC9Mli500e5AcU6" height = 90% width = 90%>
